```python
# apply() - row - wise

def discount_label(x):
    if x == 0:
        return "No Discount"
    elif x <= 0.10:
        return "Low Discount"
    else:
        return "High Discount"

df["DiscountCategory"] = df["DiscountPct"].apply(discount_label)
```

```python
df.head()
```

```python
# map() - maps the specified value to the given value

city_region = {
    "Pune" : "West",
    "Mumbai" : "West",
    "Delhi" : "North",
    "Bangalore" : "South",
    "Hyderabad" : "South",
    "Chennai" : "South"}

df["Region"] = df["City"].map(city_region)
```

```python
# Group by in Pandas

# groupby(Grouping_column)[MetricColumn].AggregateFunction
"""
Aggregate Functions :
1. mean() - Average
2. min()  - minimum value
3. sum()  - Total
4. max()  - maximum value
5. count() - TotalCount
"""

df.groupby("City")["TotalSale"].mean()
```

```python
# Multiple Aggregations
df.groupby("City")["TotalSale"].agg(["mean","sum","min","max","count"])
```

```python
# Named Aggregations
df.groupby("Category").agg(AverageTotalRevenue= ("ExactSaleAmount","mean"),
                           TotalRevenue = ("ExactSaleAmount","sum"),
                           OrderCount = ("OrderID","count"))
```

```python
df["Year"]=df["OrderDate"].dt.year                # .dt.year extracts year from the mentioned date
df["Month"]=df["OrderDate"].dt.month              # .dt.month extracts month from the mentioned date
df["Day"]=df["OrderDate"].dt.day                  # .dt.day extracts day from the mentioned date
df["MonthName"]=df["OrderDate"].dt.month_name()   # .dt.month_name() extracts month name from the mentioned date
df["DayName"]=df["OrderDate"].dt.day_name()       # .dt.day_name() extracts day name from the mentioned date
df["Quarter"]=df["OrderDate"].dt.quarter          # .dt.quarter extracts quarter from the mentioned date
```
